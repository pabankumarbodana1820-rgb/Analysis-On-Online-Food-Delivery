# Outliers Analysis

# Using IQR Method

# Using BOXPLOT

```python
df1 = df.select_dtypes(include = np.number)
df1
```

```python
for i in df1:
    q1 = df1[i].quantile(0.25)
    q2 = df1[i].quantile(0.50)
    q3 = df1[i].quantile(0.75)
```

```python
IQR = q3 - q1
IQR
```

```python
lw = q1-1.5*IQR
uw = q3 + 1.5*IQR
```

```python
outliers = ((df1[i]<lw)&(df1[i]>uw))
```

```python
print(outliers.sum())
```
